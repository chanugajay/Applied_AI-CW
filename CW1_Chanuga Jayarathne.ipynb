# **Part A - Application area review**
Object and image recognition represents a foundational domain within computer vision, leveraging Artificial Intelligence (AI) to interpret and evaluate visual data. Advancements in AI—particularly through deep learning—have revolutionized this field, with significant impacts across diverse sectors, including healthcare, autonomous transportation, agriculture, and retail. The introduction of Convolutional Neural Networks (CNNs) has notably expanded the capabilities of image recognition systems, enabling more accurate and scalable solutions.

Historically, image recognition systems relied on manually engineered features and traditional machine learning techniques. Methods such as Scale-Invariant Feature Transform (SIFT) and Histogram of Oriented Gradients (HOG) were commonly used for extracting distinctive features from images. However, these early approaches often struggled with generalization across diverse datasets, limiting their effectiveness in complex or varied real-world scenarios.

The emergence of deep learning, especially CNNs, marked a paradigm shift in visual recognition. This was most prominently demonstrated in the ImageNet Large Scale Visual Recognition Challenge (ILSVRC), where CNNs outperformed conventional methods by a wide margin (Krizhevsky, Sutskever, & Hinton, 2017). Subsequent developments introduced more sophisticated architectures such as VGG, Inception, and ResNet. For instance, ResNet’s introduction of residual connections addressed the vanishing gradient problem, facilitating the training of deeper networks and improving overall accuracy.

AI’s influence is particularly evident in the medical field, where CNNs support the automatic detection and classification of diseases from medical imagery. Esteva et al. (2017) demonstrated that deep learning models could achieve dermatologist-level accuracy in classifying skin cancer, showcasing AI's potential in diagnostic applications. Similarly, in radiology, CNNs aid in identifying anomalies within X-rays and MRIs, enhancing diagnostic efficiency and reliability.

In the automotive sector, object detection and recognition are crucial for enabling autonomous navigation. Algorithms such as YOLO (You Only Look Once) and Faster R-CNN allow vehicles to detect and respond to pedestrians, vehicles, and traffic signs in real time (Redmon et al., 2015). These technologies form the backbone of systems deployed in autonomous vehicles like Tesla’s Autopilot and Waymo’s self-driving platforms.

Retail and e-commerce also benefit significantly from AI-powered image recognition. Visual search tools used by platforms like Pinterest and Amazon enable users to find products through images rather than text. Additionally, object recognition systems track consumer behavior in-store, optimizing store layouts and enhancing personalized marketing strategies.

In agriculture, AI contributes to precision farming by monitoring crop health, identifying diseases or pests, and estimating yields. CNN-based systems can analyze drone imagery to assess plant conditions and guide timely interventions. Livestock management also benefits through automated animal counting and monitoring.

Security and surveillance systems leverage object recognition for real-time threat detection. AI-enhanced CCTV systems can identify unauthorized access, detect weapons, or track suspicious activity, thereby improving public safety and emergency response.

Recent research in this domain has explored overcoming core challenges such as generalization to unseen objects and reducing computational complexity. Zero-shot learning (ZSL) enables models to recognize novel objects using semantic associations without prior exposure to those classes. For example, scene understanding has been enhanced by integrating textual and visual modalities in ZSL, improving object recognition performance. Similarly, multimodal learning—combining data from various sources, such as images and natural language—has gained prominence for its capacity to deepen contextual understanding (Baltrusaitis, Ahuja, & Morency, 2019). Moreover, transfer learning has democratized access to high-performing models by allowing developers to fine-tune pre-trained networks on task-specific datasets, reducing computational demands.

Despite notable advancements, challenges remain. High-quality, labeled datasets are essential for training robust models, and data scarcity can hinder model performance. Furthermore, achieving real-time inference requires substantial computational power, posing limitations for deployment in resource-constrained environments.

Looking ahead, future research will likely focus on enhancing model efficiency and reducing reliance on extensive labeled datasets. Techniques such as semi-supervised learning, transfer learning, and integration with edge computing may offer solutions by enabling more responsive, lightweight models capable of operating in real-time with reduced latency and infrastructure demands.

# **Part B - Compare and evaluate AI techniques**


**1.   Convolutional Neural Networks (CNN)** <br/>

<p> Specifically, CNN is a type of deep learning that is designed to work on grid-data which includes; images. They used convolutional operation to obtain the spatial features from images and then make classification or prediction based on patterns. CNNs consist of three main components: The three categories of layers are convolutional layers, pooling layers and fully connected layers.

The convolutional layer puts masks (filters or kernels) on small parts of an input image and allows the CNN to discover spatial hierarchies of such features in the input image like edges, shapes, and texture. These filters move across the input image carrying out dot products to generate feature maps, which identifies important regions. Pooling layers come next to bring down the spatial dimensions of the feature maps but together with the goal of reducing computational requirements. These are max pooling where the maximum of the region is chosen and average pooling where average value is calculated. Last but not the least, fully connected layers transform extracted spatial features into a class probability space to make the network's outputs.

Unlike most CNNs, they are suitable for image classification because they enable the reduction of the number of parameters compared to fully connected networks. For instance, we have CIFAR-10, a 32x32x3 image with three RGB channels, which can be translated to have 3,072 features to be processed and this may be costly using only dense layers. CNNs overcome this by relying on several characteristics that make the model scalable in the number of datasets; One of them is the sharing of parameters in convolutional layers. </p>

**Strengths:**

*   Effective at capturing spatial hierarchies in images.
*   Works well for datasets with relatively small image sizes.
*   Reduces the number of parameters compared to fully connected networks.

**Weaknesses:**

*   Limited to 2D spatial data (not ideal for temporal or volumetric data).
*   Prone to overfitting on small datasets if regularization is not applied.
*   Requires careful tuning of hyperparameters (e.g., number of filters, kernel size).

**Advantages:**

*   Highly optimized in frameworks (e.g., TensorFlow, PyTorch).
*   Easy to integrate with techniques like data augmentation and dropout for regularization.

**Disadvantages:**

*   May struggle with very large datasets or datasets requiring deeper architectures for higher accuracy.


**2.   Convolution 3D**
<p> Conv3D is an extension of the typical CNNs that have been proposed for managing 3D data or sequence through time including films or medical. Unlike 2D filters used by Conv2D, Conv3D utilizes 3D filters to obtain spatial and temporal features at the same time. Every 3D convolution performs a computation that produces a feature map, which retains information from different frames or depths.

Conv3D for the same reason are very successful for video classification where temporal information is essential. For instance, in human activity recognition, Conv3D models capture spatial arrangements within the frame and temporal arrangements between the frames. Likewise in medical image processing, Conv3D is able to work on volumetric scans such as CT or MRI scans where depth information is pivotal to diagnosis.

However, Conv3D models is costlier in terms of computational resources and consumes vastly more memory than 2D CNNs. They also require large datasets in order to avoid overfitting as the inclusion of an additional dimension in a model process increases it. In the cases where there is minimal temporal or depth context information available, Conv3D models are not effective and could be a waste of computational resources. </p>

**Strengths:**

*   Temporal/Depth Data: Processes temporal or volumetric data effectively.
*   Captures Temporal Dynamics: Ideal for video or time-series tasks.

**Weaknesses:**

*   Not suitable for purely 2D data like CIFAR-10.
*   Computationally expensive and memory-intensive.

**Advantages:**

*   Captures spatial and temporal features simultaneously.

**Disadvantages:**

*   Requires large datasets and powerful hardware. <br/>

**3.   Dense Convolutional Networks** <br/>
<p> Indeed, DenseNet innovatively applies dense connections in the feed-forward nature of the deep learning layers. In DenseNet, the output of each layer serves as an input to all subsequent layers but in a CNN, only direct preceding layer’s output feeds into the subsequent layer. This architecture enables the greatest reuse of features throughout the layers, and facilitates a fast flowing gradient during backpropagation.

Composed of batch normalization, ReLU activation and convolutional operation, DenseNet layers are extremely deep. Implementing features maps from earlier layers to the next layer simplifies the task of the subsequent layers and reduces the number of features to be relearned. Further, DenseNet is proficient in terms of the number of parameters where unlike ordinary CNNs in which once computed and discarded, intermediate feature maps are recycled here. At the same time, DenseNet also resolves the vanishing gradient problem that affects most deep networks.

This architecture performs particularly well on problems that involve deep architectures, including object detection or fine-grained image classifiers. DenseNet has brought new records breaking its predecessor on large datasets such as ImageNet and CIFAR10 while improving on the basic CNN architecture. Nevertheless, by reducing the cardinality of the features, DenseNet expands computational load and memory consumption where huge data is involved. </p>

**Strengths:**

*   Feature Reusability: Promotes feature reuse across layers, reducing redundancy.
*   Fewer Parameters: Despite being deep, DenseNet is computationally efficient due to feature reuse.
*   Improved Gradient Flow: Easier training due to direct connections between layers.

**Weaknesses:**

*   Computationally expensive for datasets with very large images or high resolution.
*   Dense connections can lead to high memory usage.

**Advantages:**

*   Reduces overfitting due to efficient feature propagation
*   Works well for small datasets as features from earlier layers are reused.

**Disadvantages:**

*   Requires more GPU memory due to dense connectivity.

# **Part D - Software Testing.**

The testing methods for Convolutional Neural Networks (CNNs) encompass various strategies aimed at validating the implementation and performance of these models.<br/>

**Unit Testing** it's focuses on individual components, such as data preprocessing and model architecture. This includes verifying that image normalization correctly scales pixel values and ensuring that data augmentation techniques like rotation and flipping operate as intended. Additionally, it is crucial to confirm that the CNN architecture aligns with the design specifications, including layer count and activation functions.<br/>

**Integration Testing** evaluates the interaction between multiple components, particularly the data pipeline and model training. This involves ensuring that preprocessed data is accurately fed into the model and that the optimizer, loss function, and backpropagation processes work together without issues. It is also essential to check for proper utilization of validation and test datasets to prevent data leakage.<br/>

**Performance Testing** assesses the model's accuracy, loss, and computational efficiency. Key indicators include monitoring training accuracy and loss over epochs to ensure they reflect expected trends—namely, a decrease in loss and an increase in accuracy. The final test accuracy should typically align with benchmarks, such as achieving around 70-80% accuracy on datasets like CIFAR-10.

**Regression Testing** ensures that modifications to the model do not negatively impact its performance. After any changes, it is necessary to re-evaluate the model on the same dataset and compare metrics like accuracy, precision, recall, and F1-score against a baseline.<br/>

In analyzing results from CNN training, one observes critical metrics such as training accuracy and loss for each epoch. Ideally, training loss should decrease while accuracy increases consistently. If a high training accuracy (e.g., >85%) is accompanied by low loss, this indicates effective learning. Validation metrics are equally important; a significant divergence between training and validation accuracy can signal overfitting.
After training, test accuracy is evaluated against expected results—basic CNNs typically achieve 70-80% on CIFAR-10. The confusion matrix further aids in assessing classification performance across different classes; higher diagonal values denote correct classifications while imbalances may indicate areas needing improvement.
Overall, these testing methods are vital for ensuring that CNN implementations are robust, efficient, and capable of generalizing well to unseen data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import classification_report, confusion_matrix
```

LOADING DATA


```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

DATA VISUALIZATION

```python
labels = ['airplane', 'automobile', 'bird', 'cat', 'deer',
          'dog', 'frog', 'horse', 'ship', 'truck']

# Let's view more images in a grid format
# Define the dimensions of the plot grid
W_grid = 10
L_grid = 10

# fig, axes = plt.subplots(L_grid, W_grid)
# subplot return the figure object and axes object
# we can use the axes object to plot specific figures at various locations

fig, axes = plt.subplots(L_grid, W_grid, figsize = (17,17))

axes = axes.ravel() # flaten the 15 x 15 matrix into 225 array

n_train = len(X_train) # get the length of the train dataset

# Select a random number from 0 to n_train
for i in np.arange(0, W_grid * L_grid): # create evenly spaces variables

    # Select a random number
    index = np.random.randint(0, n_train)
    # read and display an image with the selected index
    axes[i].imshow(X_train[index,1:])
    label_index = int(y_train[index])
    axes[i].set_title(labels[label_index], fontsize = 8)
    axes[i].axis('off')

plt.subplots_adjust(hspace=0.4)
```

```python
classes_name = ['Airplane', 'Automobile', 'Bird', 'Cat', 'Deer', 'Dog', 'Frog', 'Horse', 'Ship', 'Truck']

classes, counts = np.unique(y_train, return_counts=True)
plt.barh(classes_name, counts)
plt.title('Class distribution in training set')
```

```python
classes, counts = np.unique(y_test, return_counts=True)
plt.barh(classes_name, counts)
plt.title('Class distribution in testing set')
```

DATA PREPROCESSING

```python
# Scale the data
X_train = X_train / 255.0
X_test = X_test / 255.0

# Transform target variable into one-hotencoding
y_cat_train = to_categorical(y_train, 10)
y_cat_test = to_categorical(y_test, 10)
```

```python
y_cat_train
```

MODEL BUILDING

```python
INPUT_SHAPE = (32, 32, 3)
KERNEL_SIZE = (3, 3)
model = Sequential()

# Convolutional Layer
model.add(Conv2D(filters=32, kernel_size=KERNEL_SIZE, input_shape=INPUT_SHAPE, activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(Conv2D(filters=32, kernel_size=KERNEL_SIZE, input_shape=INPUT_SHAPE, activation='relu', padding='same'))
model.add(BatchNormalization())
# Pooling layer
model.add(MaxPool2D(pool_size=(2, 2)))
# Dropout layers
model.add(Dropout(0.25))

model.add(Conv2D(filters=64, kernel_size=KERNEL_SIZE, input_shape=INPUT_SHAPE, activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(Conv2D(filters=64, kernel_size=KERNEL_SIZE, input_shape=INPUT_SHAPE, activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(filters=128, kernel_size=KERNEL_SIZE, input_shape=INPUT_SHAPE, activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(Conv2D(filters=128, kernel_size=KERNEL_SIZE, input_shape=INPUT_SHAPE, activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
# model.add(Dropout(0.2))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(10, activation='softmax'))

METRICS = [
    'accuracy',
    tf.keras.metrics.Precision(name='precision'),
    tf.keras.metrics.Recall(name='recall')
]
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=METRICS)
```

```python
model.summary()
```

EARLY STOPPING


```python
early_stop = EarlyStopping(monitor='val_loss', patience=2)
```

DATA AGUMENTATIONS

```python
batch_size = 32
data_generator = ImageDataGenerator(width_shift_range=0.1, height_shift_range=0.1, horizontal_flip=True)
train_generator = data_generator.flow(X_train, y_cat_train, batch_size)
steps_per_epoch = X_train.shape[0] // batch_size

r = model.fit(train_generator,
              epochs=5,
              steps_per_epoch=steps_per_epoch,
              validation_data=(X_test, y_cat_test),
#               callbacks=[early_stop],
#               batch_size=batch_size,
             )
```

EVALUATION OF DATA

```python
#accuracy
plt.figure(figsize=(12, 16))

plt.subplot(4, 2, 1)
plt.plot(r.history['loss'], label='Loss')
plt.plot(r.history['val_loss'], label='val_Loss')
plt.title('Loss Function Evolution')
plt.legend()

plt.subplot(4, 2, 2)
plt.plot(r.history['accuracy'], label='accuracy')
plt.plot(r.history['val_accuracy'], label='val_accuracy')
plt.title('Accuracy Function Evolution')
plt.legend()

plt.subplot(4, 2, 3)
plt.plot(r.history['precision'], label='precision')
plt.plot(r.history['val_precision'], label='val_precision')
plt.title('Precision Function Evolution')
plt.legend()

plt.subplot(4, 2, 4)
plt.plot(r.history['recall'], label='recall')
plt.plot(r.history['val_recall'], label='val_recall')
plt.title('Recall Function Evolution')
plt.legend()
```

```python
evaluation = model.evaluate(X_test, y_cat_test)
print(f'Test Accuracy : {evaluation[1] * 100:.2f}%')

y_pred = model.predict(X_test)
y_pred = np.argmax(y_pred, axis=1)
cm = confusion_matrix(y_test, y_pred)


disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=labels)


# NOTE: Fill all variables here with default values of the plot_confusion_matrix
fig, ax = plt.subplots(figsize=(10, 10))
disp = disp.plot(xticks_rotation='vertical', ax=ax,cmap='summer')

plt.show()
```

```python
print(classification_report(y_test, y_pred))
```

```python
def plot_image(i, predictions_array, true_label, img):
    predictions_array, true_label, img = predictions_array, true_label[i], img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])

    plt.imshow(img, cmap=plt.cm.binary)

    predicted_label = np.argmax(predictions_array)
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'

    plt.xlabel(f"{labels[int(predicted_label)]} {100*np.max(predictions_array):2.0f}% ({labels[int(true_label)]})",
               color=color)

def plot_value_array(i, predictions_array, true_label):
    predictions_array, true_label = predictions_array, int(true_label[i])
    plt.grid(False)
    plt.xticks(range(10))
    plt.yticks([])
    thisplot = plt.bar(range(10), predictions_array, color="#777777")
    plt.ylim([0, 1])
    predicted_label = np.argmax(predictions_array)

    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('blue')
```

```python
predictions = model.predict(X_test)

# Plot the first X test images, their predicted labels, and the true labels.
# Color correct predictions in blue and incorrect predictions in red.
num_rows = 8
num_cols = 5
num_images = num_rows * num_cols
plt.figure(figsize=(2 * 2 * num_cols, 2 * num_rows))
for i in range(num_images):
    plt.subplot(num_rows, 2 * num_cols, 2 * i + 1)
    plot_image(i, predictions[i], y_test, X_test)
    plt.subplot(num_rows, 2*num_cols, 2*i+2)
    plot_value_array(i, predictions[i], y_test)
plt.tight_layout()
plt.show()
```
